```python
# Import de bibliotecas basicas para analise de Dados - Pandas & Numpy
import numpy as np
import pandas as pd

# Import da Biblioteca Seaborn apenas para auxilixar com alguns DataSets de interesse
import seaborn as sns

# Carregar Dataset do Titanic
titanic = sns.load_dataset("titanic")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Supondo que 'titanic' é o DataFrame que você possui
# Certifique-se de ter importado os módulos necessários e carregado os dados corretamente

# Crie um DataFrame
data = titanic
df = pd.DataFrame(data)

# Configurar o tamanho do gráfico
plt.figure(figsize=(10, 6))

# Criar o countplot usando Matplotlib
survived_by_class = df.groupby(['pclass', 'survived']).size().unstack()
survived_by_class.plot(kind='bar', stacked=True, color=['#FF7171', '#71C5FF'])

# Adicionar títulos e rótulos aos eixos
plt.title('Contagem de Sobreviventes por Classe', fontsize=16)
plt.xlabel('Classe', fontsize=14)
plt.ylabel('Contagem', fontsize=14)

# Adicionar uma legenda
plt.legend(title='Sobrevivente', labels=['Não Sobreviveu', 'Sobreviveu'])

# Exibir o gráfico
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Se 'titanic' é um dicionário, converta-o para DataFrame
titanic = pd.DataFrame(titanic)

# Criar o gráfico de barras empilhadas usando Matplotlib
plt.figure(figsize=(13, 10))

# Agrupar por 'pclass', 'survived' e calcular a média da idade para cada grupo
grouped_data = titanic.groupby(['pclass', 'survived'])['age'].mean().unstack()

# Plotar o gráfico de barras empilhadas
grouped_data.plot(kind='bar', stacked=True, colormap='viridis')

# Adicionar rótulos e título
plt.xlabel('Classe')
plt.ylabel('Média de Idade')
plt.title('Média de Idade por Classe e Status de Sobrevivência')

# Mostrar o gráfico
plt.show()
```

Usando Seaborn

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

data = titanic


# Configurar o estilo e o contexto do Seaborn
sns.set(style="whitegrid")
sns.set_context("notebook", font_scale=1.2)

# Criar o gráfico Countplot usando Seaborn
plt.figure(figsize=(12, 8))
count_plot = sns.countplot(x='pclass', hue='survived', data=titanic, palette='Set1', hue_order=[0, 1], dodge=True)

# Adicionar rótulos e título
count_plot.set(xlabel='Classe', ylabel='Contagem', title='Contagem de sobreviventes e não sobreviventes por classe')

# Adicionar legenda
plt.legend(title='Sobreviveu | Não sobreviveu'
, labels=['No', 'Yes'])
```

Usando Plotly

```python
import plotly.express as px
import pandas as pd

# Seu DataFrame
data = titanic


df = pd.DataFrame(data)

# Gerar o gráfico usando Plotly Express
fig = px.scatter(df, x='pclass', y='age', color='sex', size='survived',
                 labels={'pclass': 'Pclass', 'age': 'Age'},
                 title='Survival Based on Pclass, Sex, and Age',
                 hover_data=['survived', 'who', 'fare'])

# Mostrar o gráfico
fig.show()
```
